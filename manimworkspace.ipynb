```python
%%manim SquareToCircle 

class SquareToCircle(Scene):
    def construct(self):
        circle = Circle()  # create a circle
        circle.set_fill(PINK, opacity=0.5)  # set color and transparency

        square = Square()  # create a square
        square.rotate(PI / 4)  # rotate a certain amount

        self.play(Create(square))  # animate the creation of the square
        self.play(Transform(square, circle))  # interpolate the square into the circle
        self.play(FadeOut(square)) 
```

```python
%%manim integral
class integral(GraphScene):
  CONFIG={
      'y_max':8,
      'y_axis_height':5
  }

# defining the graph of the function for which we need the area

  def construct(self):
    self.show_function_graph()
  
  def show_function_graph(self):
    self.setup_axes(animate=True)

    def func(x):
      return -x**2 + 3*x +1
    
    graph = self.get_graph(func,x_min=0.2,x_max=9)
    graph.set_color(YELLOW)
    self.play(Create(graph), run_time=3)
    self.wait(1)

# defining the rectangle function

    def rect(x):
      return x

    recta = self.get_graph(rect,x_min=-1,x_max=5)

# config dictionary for the rectangles 

    kwargs = {
        'x_min':2,
        'x_max':8,
        'fill_opacity':0.75,
        'stroke_width':0.25,
    }

# getting the rectangles

    self.graph = graph
    iterations = 6

# for the starting rectangles maybe

    self.rect_list = self.get_riemann_rectangles_list(
        graph, iterations, start_color=PURPLE,end_color=ORANGE,**kwargs
    )

# for the ending rectangles maybe

    flat_rects = self.get_riemann_rectangles(
        self.get_graph(lambda x: 0), dx = 0.7, start_color = invert_color(PURPLE), end_color = invert_color(ORANGE), **kwargs
    )

# for the first iteration

    rects = self.rect_list[0]
    self.transform_between_riemann_rects(
        flat_rects,rects,
        replace_mobject_with_target_in_scene = True, run_time = 2
    )

# for animating between diffrent iterations

    for j in range(1,6):
      self.transform_between_riemann_rects(
          self.rect_list[j-1],self.rect_list[j], dx=1,
          replace_mobject_with_target_in_scene= True,
          run_time=2
      )

## for the text

    sub = Tex('Riemann Sums').set_color(BLUE)
    sub.next_to(graph, UP)
    self.play(Write(sub))
```

```python
%%manim ani1

class ani1(Scene):
  def construct(self):
    box = Rectangle(stroke_color = GREEN, stroke_opacity=0.7,
                    fill_color = RED, fill_opacity = 0.5, height=1,width=1
                    )
    self.add(box)
    self.play(box.animate.shift(RIGHT*2), run_time=2)
    self.play(box.animate.shift(UP*3), run_time=2)
    self.play(box.animate.shift(DOWN*5+LEFT*5), RUN_TIME=2)
    self.play(box.animate.shift(UP*1+RIGHT*2), run_time=2)
```

```python
%%manim ani5

class ani5(VectorScene):
  def construct(self):
    plane = NumberPlane(x_range=[-3,3], y_range=[-4,14],y_length=7,
                        x_length=6).add_coordinates()
    
    graph1 = plane.get_graph(lambda x: x**2, x_range=[-3,3], color=RED)
    graph1_lab = (
        MathTex('f(x)={x}^{2}').next_to(graph1,UR,buff=0.2).set_color(RED).scale(0.8)
    )

    c = ValueTracker(-4)

    graph2 = always_redraw(
        lambda: plane.get_graph(
            lambda x: x**2 + c.get_value(), x_range=[-3,3], color = YELLOW
        )   
    )

    graph2_lab = always_redraw(
        lambda : MathTex('f(x)={x}^{2}+c').next_to(graph2,UR,buff=0.2).set_color(YELLOW).scale(0.8)
    )

    k = ValueTracker(-3)
    dot1 = always_redraw(
        lambda: Dot().move_to(
            plane.coords_to_point(
                k.get_value, graph1.underlying_function(k.get_value())
            )
        )
    )

    slope1 = always_redraw(
        lambda: plane.get_secant_slope_group(
            x=k.get_value(), graph=graph1, dx=0.01,secant_line_length=5
        )
    )

    slope2 = always_redraw(
        lambda: plane.get_secant_slope_group(
            x=k.get_value(), graph=graph2, dx=0.01,secant_line_length=5
        )
    )

    dot1 = always_redraw(
        lambda: Dot().move_to(
            plane.coords_to_point(
                k.get_value, graph2.underlying_function(k.get_value())
            )
        )
    )

    self.play(
        LaggedStart(DrawBorderThenFill(plane), Create(graph1), Create(graph2)),
        run_time=5,
        lag_ratio=1
    )

    self.add(slope1, slope2, dot1,dot2,graph1_lab, graph2_lab)

    self.play(
        k.animate.set_value(0), c.animate.set_value(2), run_time=5, rate_func=linear
    )

    self.play(
        k.animate.set_value(3), c.animate.set_value(-2), run_time=5, rate_func=linea
    )

    self.wait()
```

```python
%%manim slope

class slope(Scene):
  def construct(self):
    plane = NumberPlane(x_range=[-10,10], y_range=[-4,8],y_length=10,
                        x_length=15).add_coordinates()

    graph = plane.get_graph(lambda x: x**2, x_range=[-3,3], color=RED)

    tanline = TangentLine(vmob = graph, alpha=0.125)
    self.play(Create(plane), run_time = 2)
    self.play(Create(graph), run_time =2)
    self.play(Create(tanline))
    self.wait()
```

```python
%%manim Derivatives

class Derivatives(Scene):
  def construct(self):

    k =  ValueTracker(-3)

    plane1 = (
        NumberPlane(x_range=[-3,4,1], x_length=5, y_range=[-8,9,2],y_length=5)
        .add_coordinates()
        .shift(LEFT*3.5)
    )

    func1 = plane1.get_graph(
        lambda x: (1 / 3) * x ** 3, x_range=[-3,3], color=RED_C
    )

    moving_slope = always_redraw(
        lambda : plane1.get_secant_slope_group(
            x=k.get_value(),
            graph = func1,
            dx=0.5,
            secant_line_length=4,
            secant_line_color=YELLOW
        )
    )

    #dot = always_redraw(
        #lambda: Dot.move_to(
            #plane1.c2p(k.get_value(), func1.underlying_function(k.get_value()))
        #)
    #)

    self.play(LaggedStart(
        DrawBorderThenFill(plane1),
        Create(func1),
        run_time=5
    ))
    self.add(moving_slope)
    self.play(k.animate.set_value(6), run_time=15, rate_func=linear)
    self.wait()
```

```python
%%manim VectorArrow
class VectorArrow(Scene):
    def construct(self):
        dot = Dot(ORIGIN)
        arrow = Arrow(ORIGIN, [1, 0,0], buff=0)
        arrow2 = Arrow(ORIGIN,[0, 1,0], buff=0 )
        numberplane = NumberPlane()
        tip_text = Text('CHOC(1, 0)',font_size = 25).next_to(arrow.get_end(), (DOWN*2)*(RIGHT))
        tip_2 = Text('CLOTHES(0,1)',font_size = 25).next_to(arrow2.get_end(),LEFT)
        self.play(Create(numberplane))
        self.play(Create(dot))
        self.play(Create(arrow))
        self.play(Create(tip_text))
        self.play(Create(arrow2))
        self.play(Create(tip_2))
```
