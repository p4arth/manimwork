```python
def factorial(n):
  if n <= 1:
    return 1
  else:
    return n * factorial(n-1)
```

```python
def GetBinomial(n,p):
  l = []
  for i in range(n+1):
    x = ((factorial(n))/((factorial(i) * factorial(n-i)))) * (p**i) * ((1-p)**(n-i))
    l.append(x)
  return l
def GetXCoordinates(n):
  l2 = []
  for i in range(n+1):
    l2.append(i)
  return l2
```

```python
%%manim SquareToCircle 

class SquareToCircle(Scene):
    def construct(self):
        circle = Circle()  # create a circle
        circle.set_fill(PINK, opacity=0.5)  # set color and transparency

        square = Square()  # create a square
        square.rotate(PI / 4)  # rotate a certain amount

        self.play(Create(square))  # animate the creation of the square
        self.play(Transform(square, circle))  # interpolate the square into the circle
        self.play(FadeOut(square)) 
```

```python
%%manim ani5

class ani5(VectorScene):
  def construct(self):
    plane = NumberPlane(x_range=[-3,3], y_range=[-4,14],y_length=7,
                        x_length=6).add_coordinates()
    
    graph1 = plane.get_graph(lambda x: x**2, x_range=[-3,3], color=RED)
    graph1_lab = (
        MathTex('f(x)={x}^{2}').next_to(graph1,UR,buff=0.2).set_color(RED).scale(0.8)
    )

    c = ValueTracker(-4)

    graph2 = always_redraw(
        lambda: plane.get_graph(
            lambda x: x**2 + c.get_value(), x_range=[-3,3], color = YELLOW
        )   
    )

    graph2_lab = always_redraw(
        lambda : MathTex('f(x)={x}^{2}+c').next_to(graph2,UR,buff=0.2).set_color(YELLOW).scale(0.8)
    )

    k = ValueTracker(-3)
    dot1 = always_redraw(
        lambda: Dot().move_to(
            plane.coords_to_point(
                k.get_value, graph1.underlying_function(k.get_value())
            )
        )
    )

    slope1 = always_redraw(
        lambda: plane.get_secant_slope_group(
            x=k.get_value(), graph=graph1, dx=0.01,secant_line_length=5
        )
    )

    slope2 = always_redraw(
        lambda: plane.get_secant_slope_group(
            x=k.get_value(), graph=graph2, dx=0.01,secant_line_length=5
        )
    )

    dot1 = always_redraw(
        lambda: Dot().move_to(
            plane.coords_to_point(
                k.get_value, graph2.underlying_function(k.get_value())
            )
        )
    )

    self.play(
        LaggedStart(DrawBorderThenFill(plane), Create(graph1), Create(graph2)),
        run_time=5,
        lag_ratio=1
    )

    self.add(slope1, slope2, dot1,dot2,graph1_lab, graph2_lab)

    self.play(
        k.animate.set_value(0), c.animate.set_value(2), run_time=5, rate_func=linear
    )

    self.play(
        k.animate.set_value(3), c.animate.set_value(-2), run_time=5, rate_func=linea
    )

    self.wait()
```

```python
%%manim slope

class slope(Scene):
  def construct(self):
    plane = NumberPlane(x_range=[-10,10], y_range=[-4,8],y_length=10,
                        x_length=15).add_coordinates()

    graph = plane.get_graph(lambda x: (((1/((6)*(2*3.141)))*(2.718**(-1/2(((x-5)/(6))**2))))), x_range=[-3,3], color=RED)

    tanline = TangentLine(vmob = graph, alpha=0.125)
    self.play(Create(plane), run_time = 2)
    self.play(Create(graph), run_time =2)
    self.play(Create(tanline))
    self.wait()
```

```python
%%manim VectorArrow
class VectorArrow(Scene):
    def construct(self):
        dot = Dot(ORIGIN)
        arrow = Arrow(ORIGIN, [1, 0,0], buff=0)
        arrow2 = Arrow(ORIGIN,[0, 1,0], buff=0 )
        numberplane = NumberPlane()
        tip_text = Text('CHOC(1, 0)',font_size = 25).next_to(arrow.get_end(), (DOWN*2)*(RIGHT))
        tip_2 = Text('CLOTHES(0,1)',font_size = 25).next_to(arrow2.get_end(),LEFT)
        self.play(Create(numberplane))
        self.play(Create(dot))
        self.play(Create(arrow))
        self.play(Create(tip_text))
        self.play(Create(arrow2))
        self.play(Create(tip_2))
```
